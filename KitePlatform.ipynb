```python
import sys
from email_sender import *
import os
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import JavascriptException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import UnexpectedAlertPresentException
from session import *
from time import sleep
from pathlib import Path
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import time
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC

session = Session(profile=False)
browser = session.get_browser()
dirpath = os.getcwd()
```

```python
log.info('Acessando site da Kite Platform...')
browser.get('https://kiteplatform-api.telefonica.com/')
sleep(1)
log.info('Inserindo Nome de Usuário...')
browser.find_element_by_xpath("//input[@name='user']").send_keys('usuário')
sleep(1)
log.info('Inserindo Password...')
browser.find_element_by_xpath("//input[@name='password']").send_keys('senha')
sleep(1)
log.info('Clicando em Login...')
browser.find_element_by_xpath("//button[@id='login-btn']").click()
```

```python
log.info('Abrindo o dataframe Excel...')
df = pd.read_excel('remanejamento_vivo.xlsx')
sleep(1)
for index, row in df.iterrows():
    log.info('Linha ' + str(index+1) + ' de ' + str(len(df.index)) + '.')
    log.info('MSISDN ' + str(row['msisdn']) + '.')

    log.info('Acessando Inventário SIM...')
    browser.get('https://kiteplatform-api.telefonica.com/#/sim/list')

    while True:
        try:
            log.info('Preenchendo o MSISDN...')
#             browser.find_element_by_name('searchbox').send_keys('5581982599783')
#             browser.find_element_by_name('searchtextbox').send_keys(str(row['msisdn']))
            browser.find_element_by_xpath("//input[@name='searchtextbox']").send_keys(str(row['msisdn']))
            sleep(1)
            break
        except:
            sleep(3)
            continue

    log.info('Pressionando o botão Enter...')
    # browser.find_element_by_name('searchbox').send_keys(Keys.ENTER)
    browser.find_element_by_name('searchtextbox').send_keys(Keys.ENTER)
    sleep(3)
    
    while True:
        try:
            log.info('Visualizando detalhes do SIM...')
            browser.find_element_by_xpath("//div[@class='x-grid-cell-inner x-selectable']/span").click()
            sleep(1)
            break
        except:
            sleep(3)
            continue

    while True:
        try:
            log.info('Clicando em substituir SIM...')
            browser.find_element_by_xpath("//span[text()='Substituir SIM']").click()
            sleep(1)
            break
        except:
            sleep(3)
            continue
    
    log.info('Alterando o SIM para ' + str(row['iccid']) + '.')
    log.info('Preencendo com o ICC...')
    browser.find_element_by_name("icc").send_keys(str(row['iccid']))
    
    while True:
        try:
            log.info('Clicando em proceda...')
            browser.find_element_by_xpath("//span[text()='Proceda']").click()
            sleep(1)
            break
        except:
            sleep(3)
            continue

    while True:
        try:
            log.info('Clicando em Substituir...')
            browser.find_element_by_xpath("//span[text()='Substituir']").click()
            sleep(1)
            break
        except:
            sleep(3)
            continue

    while True:
        try:
            log.info('Clicando em Terminar...')
            browser.find_element_by_xpath("//span[text()='Terminar']").click()
            sleep(1)
            break
        except:
            sleep(3)
            continue
```
